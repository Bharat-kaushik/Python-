```python
import os
import psutil

print("----------------------------------------------------------USER GUIDE--------------------------------------------------\n\n\n")
print("..............With this program you can perform the following tasks by just typing your queries............ ")
print("START APPLICATION PROGRAMS LIKE CHROME/MOZILA/EMAIL/NOTEPAD... \nCLOSE APPLICATION PROGRAMS \nOPEN ANY WEBSITE BY TYPING THE WEBSITE ADDRESS WITH OPEN/RUN KEYWORDS")
print("SHUTDOWN/RESTART THE SYSTEM(IF YOU DON'T WANT TO SHUTDOWN/RESTART YOUR SYSTEM PLEASE DON'T USE SHUTDOWN/RESTART KEYWORD IN QUERY)")
print("TO EXIT FROM PROGRAM SIMPLY WRITE-> EXTI\n\n\n")
print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^\n\n\n")


programs={
    "chrome"     :   ["google chrome","chrome"],
    "mspaint"    :   ["paint","drawing","mspaint"],
    "firefox"    :   ["firefox","mozilla","mozilla firefox"],
    "wmplayer"   :   ["windows media player","wmplayer","player","media","song"],
    "vlc"        :   ["vlc","video","vlc player","vlc media player","music"],
    "notepad"    :   ["notepad","editor","text editor","create text","text"],  
    "start cmd"  :   ["command prompt","cmd","start cmd","command","terminal"],
    "write"      :   ["wordpad","word pad","write"],
    "winword"    :   ["ms word","ms-word","microsoft word","word"],
    "excel/e"    :   ["ms excel","excel","microsoft excel"],
    "powerpnt/b" :   ["power point","powerpoint","powerpnt"],
    
    "explorer"   :   ["files","file manager","file explorer","file"],
    "taskmgr"    :   ["taskmanager","task manager","task","taskmgr"],
    "start iexplore":["internet explorer","internet","iexplore"],
    "control panel" :["control panel","control","panel"],
    "start microsoft.windows.camera:":["camera","cam","selfie","click images","take photo","photo","webcam"],
    "chrome https://mail.google.com/mail/u/0/":["email","gmail","mail"],
    "start ms-settings:":["setting","settings"]  
}

while(True):
    print("Hey! what can i do for you  - ",end=" ")
    choice=input().lower()
    find=False
    chlist=choice.strip().split(" ")
    
#-------------------------------------------- To shutdown / restart------------------------------------------------- 
    if("shutdown" in chlist or "power off" in chlist):
        signal=os.system("shutdown /s /t 1")
        if(not signal):
            print("Shutting down your pc")
        else:
            print("Something went wrong!! please try again:((")
        break
    elif("restart" in chlist or "reboot" in chlist):
        signal =os.system("shutdown /r /t 1")
        if(not signal):
            print("Restarting...")
        else:
            print("something went wrong!! Please try again:((")
        break
        
#---------------------------------- ------to open a program------------------------------------------------------
    elif("run" in choice or "start" in choice or "open" in choice or "execute" in choice):
        for item in(chlist):
            for key in programs:
                if (item in programs[key]):
                    find=True
                    print("starting ",item,"...:))")
                    os.system(key)
                    break
            if (item.endswith(".com") or item.endswith(".org") or item.endswith(".in") or item.endswith(".co.in")):
                find=True
                os.system("start iexplore "+item)
                break
    
# -----------------------------------To close a program----------------------------------------------------------
    elif("close" in choice or "terminate" in choice or "stop" in choice or "end" in choice or "ends" in choice):
        mainList=["close","terminate","end","ends","stop"]  
        # prepare a list of all the running process names to check whether process is running or not
        runningProgList=[]
        for prog in psutil.process_iter():
            runningProgList.append(prog.name().lower())
        
        arg="taskkill /f /im "
        for item in (chlist):
            if item in mainList:
                continue
            for proc in (runningProgList):
                if item in (proc):
                    arg2=arg+proc
                    check=os.system(arg2)
                    if(not check):
                        print(item," closed successfully :))")
                        find=True
                        break;
#                     else:
#                         print("Something went wrong...find your process from the following list and close using specific name")
            if(not find):
                print("\ncan't find right process...???")
                print("if you want to see all running process then enter 'yes' and choose among them")
                n=input().lower()
                if(n=='yes'):
                    print(runningProgList,"\n\n")                  
            find=True
            break;

    if("exit" in choice):
        break
    if(not find):
        print("sorry! can't find the program you entered! please try again with more specific keywords:((")
```
